import pandas as pd
import numpy as np

df=pd.read_csv('data/projectdata/report3_5.csv')

```python
df.columns
```

```python
df.info()
```

```python
#나라이름 원핫인코딩
one_c=pd.get_dummies(df['Country'])
one_c.head()
```

```python
#합치기 
ndf=pd.concat([df,one_c],axis=1)
ndf.info()
```

```python
pd.set_option('display.width',None)      # 출력화면의 너비
pd.set_option('display.max_rows',None)    # 출력할 행의 개수 한도
pd.set_option('display.max_columns',None)  # 출력할 열의 개수 한도
pd.set_option('display.max_colwidth',20) # 출력할 열의 너비
```

```python
# 행복지수로 내림차순 
ndf.sort_values(by='Life Ladder', ascending=False,inplace=True)
```

```python
# 인덱스 재정렬
ndf.reset_index(drop=True, inplace=True)  
ndf.head()
```

```python
# 행의 개수 
len(ndf)
```

```python
newlist = []
for idx in ndf.index :
    if idx <= round(len(ndf)*0.2)-1 :
        newlist += ['1']
    elif idx <= round(len(ndf)*0.4)-1 :
        newlist += ['2']
    elif idx <= round(len(ndf)*0.6)-1 :
        newlist += ['3']
    elif idx <= round(len(ndf)*0.8)-1 :
        newlist += ['4']
    else :
        newlist += ['5']
        
ndf['rank'] = newlist
print(ndf['rank'].unique())
```

```python
ndf.head()
```

```python
# 독립변수과 종속변수 나누기
X = ndf.drop(['Country','Year','Life Ladder','rank'], axis=1)
y = ndf['rank']
```

```python
# 정규화하기
from sklearn import preprocessing
X = preprocessing.StandardScaler().fit(X).transform(X)
```

```python
# train 데이터와 test 데이터로 구분
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2, random_state=4)
print('train data 개수: ', X_train.shape)   
print('test data 개수: ', X_test.shape)     
```

```python
###### 1. KNN 분류 모형 ######
##### KNN 분류 모형 - sklearn 사용
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)
y_hat = knn.predict(X_test)
print(y_hat[0:10])          
print(y_test.values[0:10])  
```

```python
# 혼동행렬
from sklearn import metrics
knn_matrix = metrics.confusion_matrix(y_test, y_hat)
print(knn_matrix)
```

```python
# 성능평가하기
knn_report = metrics.classification_report(y_test, y_hat)
print(knn_report)
```

```python
# 정확도
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
print("정확도(Accuracy) :", accuracy_score(y_test,y_hat)) 
```

```python
###### 2. SVM 분류 모형 ######

from sklearn import svm
svm_model = svm.SVC(kernel='rbf')
svm_model.fit(X_train, y_train)
y_hat= svm_model.predict(X_test)
print(y_hat[0:10])          
print(y_test.values[0:10])  
```

```python
# 혼동행렬
from sklearn import metrics
svm_matrix = metrics.confusion_matrix(y_test,y_hat)
print(svm_matrix)
```

```python
# 성능평가하기
svm_report = metrics.classification_report(y_test, y_hat)
print(svm_report)
```

```python
# 정확도
print("정확도(Accuracy) : ", accuracy_score(y_test,y_hat))
```

```python
###### 3. Decision Tree 분류 모형 #####
from sklearn import tree
tree_model = tree.DecisionTreeClassifier(criterion='entropy', max_depth=5)
tree_model.fit(X_train, y_train)
y_hat = tree_model.predict(X_test)
print(y_hat[0:10])           
print(y_test.values[0:10])   
```

```python
# 혼동행렬
tree_matrix = metrics.confusion_matrix(y_test, y_hat)
print(tree_matrix)
```

```python
# 성능평가하기
tree_report = metrics.classification_report(y_test, y_hat)
print(tree_report)
```

```python
# 정확도
print("정확도(Accuracy) : ", accuracy_score(y_test,y_hat)) 
```

```python
###### 4. 로지스틱회귀분석 ######

from sklearn.linear_model import LogisticRegression
lr = LogisticRegression(random_state=4)
lr.fit(X_train,y_train)
y_hat = lr.predict(X_test)
print(y_hat[:10])           
print(y_test.values[:10])   
```

```python
###### 5. GradientBoostingClassifier 모델 ######

from sklearn.ensemble import GradientBoostingClassifier
clf = GradientBoostingClassifier()
clf.fit(X_train,y_train)
y_hat = clf.predict(X_test)
print(y_hat[:10])      
print(y_test[:10].values)
```
