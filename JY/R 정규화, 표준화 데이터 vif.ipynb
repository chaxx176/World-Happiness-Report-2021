# 1. 데이터 가져오기 (nhappy_0814.csv : 행 재배치 된 데이터)

```r
df <- read.csv("Project/pdata/nhappy_0814.csv") 
df$country <- as.factor(df$country)
df$year <- as.factor(df$year)
df$country_num <- as.factor(df$country_num)
df$continent <- as.factor(df$continent)
str(df)
```

```r
summary(df)
```

# 2. 데이터 변환

## 2-1. minmax normalization

```r
normalize <- function(x){
  return((x-min(x))/(max(x)-min(x)))
}

dfnor <- as.data.frame(lapply(df[4:15],normalize))
dfnor[c('HS','year','country','country_num','continent')] <- df[c('HS','year','country','country_num','continent')]
summary(dfnor)
```

## 2-2. standardization

```r
dfscl <- as.data.frame(lapply(df[4:15],scale))
dfscl[c('HS','year','country','country_num','continent')] <- df[c('HS','year','country','country_num','continent')]
summary(dfscl)
```

# 3. 각 데이터 회귀식 적합 
#### * model_nor 과 model_scl 회귀진단 결과 같음

## 3-1. dfnor

```r
model_nor <-lm(HS^2 ~ country +log_gdp + social + life_exp + freedom + generosity + corruption + schooling +
             alcohol + suicide + obesity + birth + mental, data=dfnor)
summary(model_nor)
```

## 3-2. dfscl

```r
model_scl <-lm(HS^2 ~ country +log_gdp + social + life_exp + freedom + generosity + corruption + schooling +
                 alcohol + suicide + obesity + birth + mental, data=dfscl)
summary(model_scl)
```

# 4. VIF 계산

## 4-1. model_nor

```r
library(car)
vif(model_nor)  # 10 이상 : log_gdp, schooling, obesity, obesity, mental
```

## 4-2. model_scl

```r
vif(model_scl)  # 10 이상 : log_gdp, schooling, obesity, obesity, mental
```
