```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns

df = pd.read_csv('Project/pdata/happy_0814.csv')
df['country']= df['country'].astype("category")
df['year']= df['year'].astype("category")
df['country_num']= df['country_num'].astype("category")
df['continent']= df['continent'].astype("category")
ndf=df.sample(frac=1,random_state=4).reset_index(drop=True)  # random_state 안 하면 DW가 실행시마다 달라짐
ndf
```

```python
# 나라별 행복지수 평균 구하기
ndf['country']= ndf['country'].astype("object")
HS_mean_country = ndf.groupby(['country'])['HS'].mean().to_frame().sort_values(by='HS',ascending = False)
HS_mean_country
```

```python
# 상위 10개 국가
HS_mean_country.head(10).index
```

```python
# 상위 10개 국가 'HS','log_gdp','life_exp','social','schooling','mental','birth' 전체 데이터 평균값 
filter = ndf['country'].isin(['Denmark', 'Finland', 'Switzerland', 'Norway', 'Netherlands', 'Iceland',
       'Canada', 'Sweden', 'New Zealand', 'Australia'])
top10 = ndf[filter][['HS','country','log_gdp','life_exp','social','schooling','mental','birth']].set_index('country')
top10 = top10.groupby(['country']).mean().reset_index()
top10
```

```python
# 상위 10개 국가의 mental, birth 평균값 
print(top10['birth'].mean())
print(top10['mental'].mean())
```
