```python
# 데이터 가져오기

import pandas as pd

pd.set_option('display.width',None)      # 출력화면의 너비
pd.set_option('display.max_rows',None)    # 출력할 행의 개수 한도
pd.set_option('display.max_columns',None)  # 출력할 열의 개수 한도
pd.set_option('display.max_colwidth',20) # 출력할 열의 너비

df = pd.read_csv('Python/Project/pdata/report3_7.csv')
df.info()
```

```python
##### 나라별 행복지수 평균값을 계산하여 가장 높은 나라 10개와 가장 낮은 나라 10개 출력하기

# 1. 나라별 행복지수 평균값 구하기
happy_mean = df.groupby('Country')['Life Ladder'].mean().sort_values(ascending = False)
happy_mean = pd.DataFrame(happy_mean)


# 가장 행복지수가 높은 나라 10개
happiest = happy_mean.head(10)


# 가장 행복지수가 낮은 나라 10개
unhappiest = happy_mean.tail(10)


# 가장 행복지수가 높은 나라와 가장 행복지수가 낮은 나라 합치기
hap_unhap = pd.concat([happiest,unhappiest])



# 가로 막대그래프로 표현하기
import seaborn as sns
plt.figure(figsize=(15,10))
sns.barplot(data= hap_unhap , x ='Life Ladder', y = hap_unhap.index )
plt.show()
```
