# 각 컬럼별 상위, 하위 5개 국가 막대그래프

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
pd.set_option('display.width',None)      # 출력화면의 너비
pd.set_option('display.max_rows',None)    # 출력할 행의 개수 한도
pd.set_option('display.max_columns',None)  # 출력할 열의 개수 한도
pd.set_option('display.max_colwidth',20) # 출력할 열의 너비

df = pd.read_csv('Project/pdata/report_0809.csv')
df.info()
```

### 1. GDP 상위, 하위 5개 국가 

```python
GDP_mean = df.groupby('Country')['Log GDP per capita'].mean().sort_values(ascending = False)
GDP_mean = pd.DataFrame(GDP_mean)
GDP_top = GDP_mean.head(5)
GDP_bot = GDP_mean.tail(5)
GDP = pd.concat([GDP_top,GDP_bot])
plt.figure(figsize=(15,10))
sns.barplot(data= GDP , x ='Log GDP per capita', y = GDP.index )
plt.show()
```
