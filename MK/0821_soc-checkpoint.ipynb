```python

import pandas as pd 
import geopandas as gpd 
import matplotlib.colors
import pycountry
from sklearn.cluster import KMeans
import seaborn as sns
import matplotlib.pyplot as plt

background_color = "#fafafa"
#colors
low_c = '#dd4124'
high_c = '#009473'
plt.rcParams["font.family"] = "monospace"
```

```python
df= pd.read_csv('C:/py/pjt/hap2021.csv')
df.columns

df.drop(['Regional indicator','Standard error of ladder score', 'upperwhisker', 'lowerwhisker','Ladder score in Dystopia',
       'Explained by: Log GDP per capita', 'Explained by: Social support',
       'Explained by: Healthy life expectancy',
       'Explained by: Freedom to make life choices',
       'Explained by: Generosity', 'Explained by: Perceptions of corruption',
       'Dystopia + residual'],axis=1,inplace = True)
df.columns
```

```python
df.rename(columns={'Country name':'country','Ladder score':'HS', 'Logged GDP per capita':'log_gdp',
       'Social support':'social', 'Healthy life expectancy':'healthy',
       'Freedom to make life choices':'free', 'Generosity':'generosity',
       'Perceptions of corruption':'corruption'},inplace = True)
```

```python
#kmeans
kmeans = KMeans(n_clusters=3, init='k-means++', max_iter=300, random_state=4)
kmeans.fit(df[['social']])
df['kmeans']= kmeans.labels_
df.head()
```

```python
geo_temp=df

def alpha3code(column):
    CODE=[]
    for country in column:
        try:
            code=pycountry.countries.get(name=country).alpha_3
            CODE.append(code)
        except:
            CODE.append('None')
    return CODE
# create a column for code 
geo_temp['CODE']=alpha3code(geo_temp['country'])
```

```python
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))
world.columns=['pop_est', 'continent', 'name', 'CODE', 'gdp_md_est', 'geometry']
merge=pd.merge(world,geo_temp,on='CODE')
```

```python

cmap = matplotlib.colors.LinearSegmentedColormap.from_list("", ['red','green','blue'])
ax = world.plot(figsize=(20,15), linewidth=0.25, edgecolor=background_color, color='lightgray')
ax.axis('off')
ax.set_facecolor(background_color)
merge.plot(column='kmeans',figsize=(20, 15),legend=False,cmap=cmap,ax=ax)


ax.text(-175,112,'Distribution of social',fontsize=30,fontweight='bold',color='#323232')
plt.show()
```
