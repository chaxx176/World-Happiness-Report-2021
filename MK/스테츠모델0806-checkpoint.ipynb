```python
import pandas as pd 
q1=pd.read_csv('C:/py/pjt/q1.csv')
q2=pd.read_csv('C:/py/pjt/q2.csv')
q3=pd.read_csv('C:/py/pjt/q3.csv')

x1=q1[q1.columns.difference(['Country','Grade','Life Ladder','Year','depressive disorders'])]
y1=q1['Life Ladder']

x2=q2[q2.columns.difference(['Country','Grade','Life Ladder','Year','depressive disorders'])]
y2=q2['Life Ladder']

x3=q3[q3.columns.difference(['Country','Grade','Life Ladder','Year','depressive disorders'])]
y3=q3['Life Ladder']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split\
    (x1,y1,test_size=0.2, random_state=4)

import statsmodels.api as sm
x_train=sm.add_constant(x_train)
model=sm.OLS(y_train, x_train).fit()
model.summary()
```
