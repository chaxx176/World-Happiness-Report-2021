```python
import pandas as pd 
import geopandas as gpd 
import matplotlib.colors
import pycountry
from sklearn.cluster import KMeans
import seaborn as sns
import matplotlib.pyplot as plt

background_color = "#fafafa"
# colours
low_c = '#dd4124'
high_c = '#009473'
plt.rcParams["font.family"] = "monospace"
df=pd.read_csv('C:/py/pjt/happy_2019.csv')
df.columns
```

```python
from sklearn import cluster
X = df[['life_exp']]
kmeans = cluster.KMeans(init='k-means++', n_clusters = 3, n_init = 10,random_state=4)
kmeans.fit(X)                 
cluster_label = kmeans.labels_
df['kmeans']= kmeans.labels_
print(df[df['kmeans']==0]['life_exp'].max()) # red
print(df[df['kmeans']==1]['life_exp'].max()) # blue 
print(df[df['kmeans']==2]['life_exp'].max()) # green
```

```python
print(df[df['kmeans']==0]['life_exp'].mean()) 
print(df[df['kmeans']==1]['life_exp'].mean()) 
print(df[df['kmeans']==2]['life_exp'].mean()) 
```

```python
geo_temp=df

def alpha3code(column):
    CODE=[]
    for country in column:
        try:
            code=pycountry.countries.get(name=country).alpha_3
            CODE.append(code)
        except:
            CODE.append('None')
    return CODE
# create a column for code 
geo_temp['CODE']=alpha3code(geo_temp['country'])
```

```python
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))
world.columns=['pop_est', 'continent', 'name', 'CODE', 'gdp_md_est', 'geometry']
merge=pd.merge(world,geo_temp,on='CODE')
```

```python
cmap = matplotlib.colors.LinearSegmentedColormap.from_list("", ['red','blue','green'])
ax = world.plot(figsize=(20,15), linewidth=0.25, edgecolor=background_color, color='lightgray')
ax.axis('off')
ax.set_facecolor(background_color)
merge.plot(column='kmeans',figsize=(20, 15),legend=False,cmap=cmap,ax=ax)


ax.text(-175,112,'Distribution of Healthy life expectancy 2019',fontsize=30,fontweight='bold',color='#323232')
plt.show()
```
