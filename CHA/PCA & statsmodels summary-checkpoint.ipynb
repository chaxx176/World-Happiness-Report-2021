# 1. statsmodels 로 summary 출력

```python
import pandas as pd
df3= pd.read_csv('dptest1.csv', header=0)
df3.head()
# 컬럼 이름 변경
df3.rename(columns={"Country name": "Country", "year":"Year","Life Ladder":"Happiness_Score", "Log GDP per capita":"Economy", 
                  "Social support":"Society" , "Healthy life expectancy at birth":"Health", "Freedom to make life choices":"Freedom",
                  "alcohol disorders":"Alcohol", "suicide rates":"Suicide", "Perceptions of corruption":"Corruption", "DALYs Mental disorders":"disorders"}, inplace=True)
df3.columns

from statsmodels.formula.api import ols
# DALYs Mental disorders 입력, 정규화 실행 전
model = ols("Happiness_Score~ Economy + Generosity+ Society + Health + Freedom + Corruption + schooling+ Alcohol+ Suicide + Obesity+ Fertility + disorders", data=df3).fit()
model.summary()
```

# 2. PCA

```python
import pandas as pd
df2= pd.read_csv('report3_7.csv', header=0)
df2.head()
x= df2.drop(['Country', 'Year','depressive disorders', 'Life Ladder'], axis=1).values
y= df2['Life Ladder'].values

# PCA, SVM은 정규화 한다음 해야함
from sklearn.preprocessing import StandardScaler
x= StandardScaler().fit_transform(x)

from sklearn.decomposition import PCA
pca= PCA()
pca.fit(x)
PC_score= pca.transform(x)
pca.components_ # eigen vector
pca.explained_variance_
ratio= pca.explained_variance_ratio_
ratio # 총 11개의 PCA 값이 나옴
```

```python
# 파이차트
df_v= pd.DataFrame(ratio[0:5], index=['PCA1','PCA2','PCA3','PCA4','PCA5'], columns=['pc_ratio'])
df_v.plot.pie(y='pc_ratio')
df_v
```

```python
# elbow plot
import matplotlib.pyplot as plt
ax= plt.plot(['PCA1','PCA2','PCA3','PCA4','PCA5'], ratio[0:5])
plt.ylabel('variance(%)')
plt.xlabel('PCA')
plt.title('PCA Plot')
plt.show()
```

```python
# 막대그래프
ax= plt.bar(x= range(len(ratio[0:5])), height= ratio[0:5], tick_label=['PCA1','PCA2','PCA3','PCA4','PCA5'])
plt.ylabel('Variance(%)')
plt.xlabel('PCA')
plt.title('PCA Plot')
plt.show()
```
